```python
def set_size(width_pt, fraction=1, subplots=(1, 1)):
    # Width of figure (in pts)
    fig_width_pt = width_pt * fraction
    # Convert from pt to inches
    inches_per_pt = 1 / 72.27 

    # Golden ratio to set aesthetic figure height
    golden_ratio = (5**.5 - 1) / 2

    # Figure width in inches
    fig_width_in = fig_width_pt * inches_per_pt
    # Figure height in inches
    fig_height_in = fig_width_in * golden_ratio * (subplots[0] / subplots[1])
    return (fig_width_in, fig_height_in)
```

```python
def eval_event_stats(e, experiment, model, transient_only=False, no_nans=False):
    # map features
    X, y, g = experiment.raw_transform(e, transient_only=False, no_nans=False)
    timestamps = X.index # save original timestamps
    
    nans = np.isnan(y) # save location of nan labels
    y_0 = np.where(nans, 0, y) # temporarily set nan to zero
    X, yt = experiment.transform(X, y_0)
    y = np.where(nans, np.nan, yt) # restore nans
    
    preds = model.predict_proba(X)
    return pd.DataFrame(data=preds, index=timestamps, columns=experiment._label_encoder.classes_), pd.Series(y, index=timestamps)
```

```python
def eval_event_wave(e, experiment_wave, model_wave, transient_only=False, no_nans=False):
    # map features
    X, y, g = experiment_wave.raw_transform(e, transient_only=False, no_nans=False)
    timestamps = X.index # save original timestamps
    
    nans = np.isnan(y) # save location of nan labels
    y_0 = np.where(nans, 0, y) # temporarily set nan to zero
    X, yt = experiment_wave.transform(X, y_0)
    y = np.where(nans, np.nan, yt) # restore nans
    
    preds = model_wave.predict_proba(X)
    return pd.DataFrame(data=preds, index=timestamps, columns=experiment_wave._label_encoder.classes_), pd.Series(y, index=timestamps)
```

```python
def eval_event_mixed(e, experiment_mixed, model_mixed, transient_only=False, no_nans=False):
    # map features
    X, y, g = experiment_mixed.raw_transform(e, transient_only=False, no_nans=False)
    timestamps = X.index # save original timestamps
    
    nans = np.isnan(y) # save location of nan labels
    y_0 = np.where(nans, 0, y) # temporarily set nan to zero
    X, yt = experiment_mixed.transform(X, y_0)
    y = np.where(nans, np.nan, yt) # restore nans
    
    preds = model_mixed.predict_proba(X)
    return pd.DataFrame(data=preds, index=timestamps, columns=experiment_mixed._label_encoder.classes_), pd.Series(y, index=timestamps)
```

```python
def eval_event_mixed(e, experiment_mixed, model_mixed, transient_only=False, no_nans=False):
    # map features
    X, y, g = experiment_mixed.raw_transform(e, transient_only=False, no_nans=False)
    timestamps = X.index # save original timestamps
    
    nans = np.isnan(y) # save location of nan labels
    y_0 = np.where(nans, 0, y) # temporarily set nan to zero
    X, yt = experiment_mixed.transform(X, y_0)
    y = np.where(nans, np.nan, yt) # restore nans
    
    preds = model_mixed.predict_proba(X)
    return pd.DataFrame(data=preds, index=timestamps, columns=experiment_mixed._label_encoder.classes_), pd.Series(y, index=timestamps)
```

```python
import pywt

def plot_waves(e, tag):
    tags, labels = e["tags"], e["labels"]
    t0 = tags.index[0]
    t_start = pd.to_datetime(t0, format='ISO8601')
    labels_index = pd.to_datetime(labels.index, format='ISO8601')
    # plot status as background color
    fig, ax = plt.subplots(1, 1, sharex=True,
                           figsize=set_size(345, subplots=(1.4, 1)),
                           layout="constrained")
    y_lims = (-10, 10)

    reltime = (labels_index - t_start).total_seconds()
    norm  = ax.fill_between(reltime, *y_lims, where=labels.values == 0,
                               color="green", alpha=0.3, label="Normal")
    tran  = ax.fill_between(reltime, *y_lims, where=labels.values > 100,
                               color="yellow", alpha=0.3, label="Transient")
    fault = ax.fill_between(reltime, *y_lims, where=(labels.values > 0) & (labels.values < 100),
                               color="red", alpha=0.3, label="Fault")
    ax.grid()
    bg_legend = ax.legend(handles=[norm, tran, fault], loc="lower left")

    # add lines
    lines = []
    for t in [tag]: #tags.columns:
        z = tags[t].values
        z = (z - np.nanmin(z)) / (np.nanmax(z) - np.nanmin(z)) # normalize from 0-1
        line, = ax.plot(reltime, z, label=t, lw=2)
        lines.append(line)
    
    # add wavelets
    MAX_LEVEL=10
    offset = 2**(MAX_LEVEL-1)
    pad_len = 2**MAX_LEVEL - len(z) % 2**MAX_LEVEL
    z = np.pad(z, [2**MAX_LEVEL, pad_len])
    Z = pywt.swt(z[::-1], level=10, start_level=0, norm=True, wavelet="haar")

    start_level = 5
    for i, (a, d) in enumerate(reversed(Z)):
        if i < start_level:
            continue
        line_d, = plt.plot(reltime, +d[::-1][2**MAX_LEVEL:-pad_len], color=f"C{i+1-start_level}", linestyle= ':', label=f"D{i+1}")
        line_a, = plt.plot(reltime, +a[::-1][2**MAX_LEVEL:-pad_len], color=f"C{i+1-start_level}", linestyle='--', label=f"A{i+1}")
        lines.append(line_d)
        lines.append(line_a)
    
    ax.set_title(e["file_name"])
    ax.set_yticklabels([])

    ax.legend(handles=lines, loc="upper right")
    #ax.add_artist(bg_legend)
    ax.set_ylim(-0.5, 1.05)
    ax.set_xlim(2**MAX_LEVEL, max(reltime) - 6*offset)

    return fig
```

```python
# get importances
np.random.seed(42)

skf = StratifiedKFold(5, shuffle=True)
importances = defaultdict(list)
reports = []


for train, test in skf.split(transformed_train_events, groups):
    train_set = MAEDataset.gather([transformed_train_events[i] for i in train])
    test_set =  MAEDataset.gather([transformed_train_events[i] for i in test])

    # get feature names
    feature_names = transformed_train_events[train[0]][0].columns
    Xt, yt, _ = experiment.balance(train_set.X, train_set.y, train_set.g, train_set.g_class)
    
    # remove nans
    isna = np.isnan(yt)
    Xt, yt = Xt[~isna], yt[~isna]
    
    # impute
    imputer = SimpleImputer()
    Xt = imputer.fit_transform(Xt)
    feature_names = feature_names[~np.isnan(imputer.statistics_)] # filter missing features
    scaler = StandardScaler()
    Xt = scaler.fit_transform(Xt)
    
    rfc = RandomForestClassifier(n_jobs=-1)
    rfc.fit(Xt, yt)
    for n, v in zip(feature_names, rfc.feature_importances_):
        importances[n].append(v)

    XT, yT = test_set.X, test_set.y
    isna = np.isnan(yT)
    XT, yT = XT[~isna], yT[~isna]
    
    XT = imputer.transform(XT)
    XT = scaler.transform(XT)
    
    y_pred = rfc.predict(XT)
    reports.append(classification_report(yT, y_pred, output_dict=True))
```

```python
import matplotlib.pyplot as plt
from matplotlib.path import Path
import matplotlib.patches as patches
import numpy as np

def parallelCoordinatesPlot(title, N, data, ynames, value=None):
    """
    A legend is added, if category_names is not None.

    :param title: The title of the plot.
    :param N: Number of data sets (i.e., lines).
    :param data: A list containing one array per parallel axis, each containing N data points.
    :param category: An array containing the category of each data set.
    :param category_names: Labels of the categories. Must have the same length as set(category).
    :param ynames: The labels of the parallel axes.
    :param colors: A colormap to use.
    :return:
    """

    w, h = set_size(345, fraction=1.5)
    #fig, ax = plt.subplots(1, 1, figsize=(w, w), layout="constrained")
    fig, host = plt.subplots(1, 1, figsize=(w, h), layout="constrained")

    # organize the data
    ys = data
    
    ymins = ys.min(axis=0)
    ymaxs = ys.max(axis=0)
    dys = ymaxs - ymins
    ymins -= dys * 0.05  # add 5% padding below and above
    ymaxs += dys * 0.05
    dys = ymaxs - ymins

    # transform all data to be compatible with the main axis
    zs = np.zeros_like(ys)
    zs[:, 0] = ys[:, 0]
    zs[:, 1:] = (ys[:, 1:] - ymins[1:]) / dys[1:] * dys[0] + ymins[0]

    axes = [host] + [host.twinx() for i in range(ys.shape[1] - 1)]
    for i, ax in enumerate(axes):
        ax.set_ylim(ymins[i], ymaxs[i])
        ax.spines['top'].set_visible(False)
        ax.spines['bottom'].set_visible(False)

        if ax != host:
            ax.spines['left'].set_visible(False)
            ax.yaxis.set_ticks_position('right')
            ax.spines["right"].set_position(("axes", i / (ys.shape[1] - 1)))

    host.set_xlim(0, ys.shape[1] - 1)
    host.set_xticks(range(ys.shape[1]))
    host.set_xticklabels(ynames, fontsize=14)
    host.tick_params(axis='x', which='major', pad=7)
    host.spines['right'].set_visible(False)
    host.xaxis.tick_top()
    host.set_title(title, fontsize=18)


    if value is not None:
        vmax, vmin = value.max(), value.min()
        vdelta = vmax - vmin
        vmax += vdelta*0.05
        vmin -= vdelta*0.05
        vdelta = vmax - vmin
        value = (value - vmin)/vdelta
        color = [matplotlib.colormaps["cividis"](v) for v in value]
    
    for j in range(N):
        # to just draw straight lines between the axes:
        # host.plot(range(ys.shape[1]), zs[j,:], c=colors[(category[j] - 1) % len(colors) ])

        # create bezier curves
        # for each axis, there will a control vertex at the point itself, one at 1/3rd towards the previous and one
        #   at one third towards the next axis; the first and last axis have one less control vertex
        # x-coordinate of the control vertices: at each integer (for the axes) and two inbetween
        # y-coordinate: repeat every point three times, except the first and last only twice
        n_axes = len(ys)
        verts = list(zip([x for x in np.linspace(0, n_axes - 1, n_axes * 3 - 2, endpoint=True)],
                         np.repeat(zs[j, :], 3)[1:-1]))
        # for x,y in verts: host.plot(x, y, 'go') # to show the control points of the beziers
        codes = [Path.MOVETO] + [Path.CURVE4 for _ in range(len(verts) - 1)]
        path = Path(verts, codes)
        patch = patches.PathPatch(path, facecolor='none', lw=1, edgecolor=color[j])
        host.add_patch(patch)

    for ax in axes:
        plt.setp(ax.get_yticklabels(), backgroundcolor=(1, 1, 1, 0.5),)


    # highlight top trial
    host.scatter(np.arange(len(zs[N-1])), zs[N-1], c='red', marker='x', s=150)
    
    
    cbar = fig.colorbar(
        matplotlib.cm.ScalarMappable(norm=matplotlib.cm.colors.Normalize(vmin=vmin, vmax=vmax), cmap="cividis"),
        ax=axes[-1], label="Balanced Accuracy", location="right", aspect=15*np.pi)
    #cbar.set_ticks([])
    return fig
```
